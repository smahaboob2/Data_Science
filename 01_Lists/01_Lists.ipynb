# Python List
### How to create a list?
    # empty list
    # list of integers
    # list with mixed datatypes
    # nested list

```python
# empty list
my_list = []
print(my_list)
```

```python
# list of integers
my_list = [1, 2, 3]
print(my_list)
```

```python
# list with mixed datatypes
my_list = [1, "Hello", 3.4]
print(my_list)
```

```python
# nested list
my_list = ["mouse", [8, 4, 6], ['a']]
print(my_list)
```

### How to access elements from a list?
    # List Index
    # Negative indexing

```python
my_list = ['p','r','o','b','e']
# Output: p
print(my_list[0])

# Output: o
print(my_list[2])

# Output: e
print(my_list[4])

# Error! Only integer can be used for indexing
# my_list[4.0]
# Nested List
n_list = ["Happy", [2,0,1,5]]

# Nested indexing
# Output: a
print(n_list[0][1])  

# Output: 5
print(n_list[1][3])
```

```python
# Negative indexing
my_list = ['p','r','o','b','e']
# Output: e
print(my_list[-1])
# Output: p
print(my_list[-5])
```

### How to slice lists in Python?

```python
my_list = ['p','r','o','g','r','a','m','i','z']
# elements 3rd to 5th
print(my_list[2:5])

# elements beginning to 4th
print(my_list[:-5])

# elements 6th to end
print(my_list[5:])

# elements beginning to end
print(my_list[:])
```

### How to change or add elements to a list?

```python
# mistake values
odd = [2, 4, 6, 8]

# change the 1st item    
odd[0] = 1   

# Output: [1, 4, 6, 8]
print(odd)

# change 2nd to 4th items
odd[1:4] = [3, 5, 7]  

# Output: [1, 3, 5, 7]
print(odd)  

odd = [1, 3, 5]
odd.append(7)

# Output: [1, 3, 5, 7]
print(odd)
odd.extend([9, 11, 13])

# Output: [1, 3, 5, 7, 9, 11, 13]
print(odd)

odd = [1, 3, 5]
# Output: [1, 3, 5, 9, 7, 5]
print(odd + [9, 7, 5])
#Output: ["re", "re", "re"]
print(["re"] * 3)

odd = [1, 9]
odd.insert(1,3)
# Output: [1, 3, 9] 
print(odd)
odd[2:2] = [5, 7]
# Output: [1, 3, 5, 7, 9]
print(odd)
```

### How to delete or remove elements from a list?

```python
my_list = ['p','r','o','b','l','e','m']
# delete one item
del my_list[2]

# Output: ['p', 'r', 'b', 'l', 'e', 'm']     
print(my_list)

# delete multiple items
del my_list[1:5] 

# Output: ['p', 'm']
print(my_list)

# delete entire list
del my_list 

# Error: List not defined
print(my_list)
```

```python
my_list = ['p','r','o','b','l','e','m']
my_list.remove('p')

# Output: ['r', 'o', 'b', 'l', 'e', 'm']
print(my_list)

# Output: 'o'
print(my_list.pop(1))

# Output: ['r', 'b', 'l', 'e', 'm']
print(my_list)

# Output: 'm'
print(my_list.pop())

# Output: ['r', 'b', 'l', 'e']
print(my_list)

my_list.clear()

# Output: []
print(my_list)
```

### List Comprehension: Elegant way to create new List

```python
pow2 = [2 ** x for x in range(10)]
# Output: [1, 2, 4, 8, 16, 32, 64, 128, 256, 512]
print(pow2)
```

```python
pow2 = []
for x in range(10):
   pow2.append(2 ** x)
print(pow2)
```

```python
pow2 = [2 ** x for x in range(10) if x > 5]
print(pow2)
```

```python
odd = [x for x in range(20) if x % 2 == 1]
odd
```

```python
[x+y for x in ['Python ','C '] for y in ['Language','Programming']]
```

### Other List Operations in Python
    # List Membership Test
    # Iterating Through a List

```python
# List Membership Test
my_list = ['p','r','o','b','l','e','m']
# Output: True
print('p' in my_list)
# Output: False
print('a' in my_list)
# Output: True
print('c' not in my_list)
```

```python
# Iterating Through a List
for fruit in ['apple','banana','mango']:
    print("I like",fruit)
```
